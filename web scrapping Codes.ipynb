```python
import requests
from bs4 import BeautifulSoup
import lxml
import time
import random
import csv

url_text = 'https://www.booking.com/searchresults.html?ss=Goa%2C+India&ssne=Chennai&ssne_untouched=Chennai&label=gen173nr-10CAEoggI46AdIM1gEaGyIAQGYATO4ARfIAQzYAQPoAQH4AQGIAgGoAgG4Au7DlcwGwAIB0gIkMTFiZTlkNWYtMWU1MC00NTI5LWJjZmItZTVjNWQ1YzhkYTBm2AIB4AIB&aid=304142&lang=en-us&sb=1&src_elem=sb&src=searchresults&dest_id=4127&dest_type=region&ac_position=0&ac_click_type=b&ac_langcode=en&ac_suggestion_list_length=5&search_selected=true&search_pageview_id=8fbb066ec124c9928d95818676f93274&ac_meta=GiA4ZmJiMDY2ZWMxMjRjOTkyOGQ5NTgxODY3NmY5MzI3NCAAKAEyAmVuOgNnb2FAAEoAUAA%3D&checkin=2026-02-22&checkout=2026-02-23&group_adults=2&no_rooms=1&group_children=0'

def web_scrapper2(web_url, file_name):

    print("Thank you for sharing the url!")
    print("Reading the content...")

    time.sleep(random.randint(5, 10))

    header = {
              "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36",
              "Accept-Language": "en-US,en;q=0.9",
              "Accept-Encoding": "gzip, deflate",
              "Connection": "keep-alive"
            }

    response = requests.get(web_url, headers=header)
    print("Status Code:", response.status_code)

    if response.status_code != 200:
        print("Failed to connect to website")
        return

    HTML_content = response.text
    soup = BeautifulSoup(HTML_content, 'lxml')

    hotel_divs = soup.find_all('div', role="listitem")

    with open(file_name, 'w', encoding='utf-8', newline='') as file_csv:
        writer = csv.writer(file_csv)
        writer.writerow(['hotel_name','location','price','rating','score','review','link'])

        for hotel in hotel_divs:
            try:
                hotel_name = hotel.find('div', class_="b87c397a13 a3e0b4ffd1").text.strip()
                location = hotel.find('span', class_="d823fbbeed f9b3563dd4").text.strip()
                price = hotel.find('span', class_="b87c397a13 f2f358d1de ab607752a2").text.replace('₹ ','')
                rating = hotel.find('div', class_="f63b14ab7a f546354b44 becbee2f63").text.strip()
                score = hotel.find('div', class_="f63b14ab7a dff2e52086").text.strip()
                review = hotel.find('div', class_="fff1944c52 fb14de7f14 eaa8455879").text.strip()
                link = hotel.find('a', href=True).get('href')

                writer.writerow([hotel_name, location, price, rating, score, review, link])
            except:
                continue

# ✅ THIS LINE IS CRUCIAL (starts the scraping)
web_scrapper2(url_text, 'hotel_data.csv')
```
